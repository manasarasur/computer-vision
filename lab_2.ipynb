```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

# Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read image (convert to grayscale for simplicity)
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if image is None:
    print("Error: Could not load image.")
else:
    print("Image loaded successfully!")

    # Kernel size
    ksize = 3
    pad = ksize // 2
    padded = np.pad(image, pad, mode='edge')

    # Prepare output images
    min_filtered = np.zeros_like(image)
    max_filtered = np.zeros_like(image)

    # Apply Min & Max filtering manually
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            neighborhood = padded[i:i+ksize, j:j+ksize].flatten()
            min_filtered[i, j] = np.min(neighborhood)   # Minimum value
            max_filtered[i, j] = np.max(neighborhood)   # Maximum value

    # Show results
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.title("Original")
    plt.imshow(image, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Min Filter (darkens)")
    plt.imshow(min_filtered, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Max Filter (brightens)")
    plt.imshow(max_filtered, cmap='gray')
    plt.axis("off")

    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read image (grayscale for frequency filtering)
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Could not load image.")
else:
    print("Image loaded successfully!")

    # FFT (2D Fourier Transform)
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)  # shift zero freq to center
    magnitude_spectrum = 20 * np.log(np.abs(fshift) + 1)

    # Create filters
    rows, cols = img.shape
    crow, ccol = rows//2, cols//2

    # Low-pass mask (circle around center)
    mask_low = np.zeros((rows, cols), np.uint8)
    r = 30   # radius of low-pass
    cv2.circle(mask_low, (ccol, crow), r, 1, thickness=-1)

    # High-pass mask (remove center, keep edges)
    mask_high = 1 - mask_low

    # Band-pass mask (ring)
    mask_band = np.zeros((rows, cols), np.uint8)
    r_out, r_in = 80, 30
    cv2.circle(mask_band, (ccol, crow), r_out, 1, thickness=-1)
    cv2.circle(mask_band, (ccol, crow), r_in, 0, thickness=-1)

    # Apply filters
    f_low = fshift * mask_low
    f_high = fshift * mask_high
    f_band = fshift * mask_band

    # Inverse FFT to get images back
    img_low = np.abs(np.fft.ifft2(np.fft.ifftshift(f_low)))
    img_high = np.abs(np.fft.ifft2(np.fft.ifftshift(f_high)))
    img_band = np.abs(np.fft.ifft2(np.fft.ifftshift(f_band)))

    # Show results
    plt.figure(figsize=(14,8))

    plt.subplot(2,3,1), plt.imshow(img, cmap='gray')
    plt.title("Original"), plt.axis("off")

    plt.subplot(2,3,2), plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title("FFT Spectrum"), plt.axis("off")

    plt.subplot(2,3,3), plt.imshow(img_low, cmap='gray')
    plt.title("Low-Pass (Blurred)"), plt.axis("off")

    plt.subplot(2,3,4), plt.imshow(img_high, cmap='gray')
    plt.title("High-Pass (Edges)"), plt.axis("off")

    plt.subplot(2,3,5), plt.imshow(img_band, cmap='gray')
    plt.title("Band-Pass (Mid-Details)"), plt.axis("off")

    plt.show()
```
